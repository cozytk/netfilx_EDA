value_counts

```python
for col in df.columns:
    print(col)
    print(df[col].value_counts())
```

```python
for col in df.columns:
    x = list(df[col].value_counts().head(10).keys())
    y = list(df[col].value_counts().head(10).values)
    plt.bar(x, y)
    plt.xticks(rotation=45)
    plt.show()
```
